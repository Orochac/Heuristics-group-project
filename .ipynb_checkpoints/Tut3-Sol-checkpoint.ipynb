The Traveling Salesperson Problem is a classic combinatorial optimization problem in computer science and mathematics. 

The problem can be defined as follows: given a list of cities and the distances between each pair of cities, find the shortest possible route that visits each city exactly once and returns to the starting city.

In other words, the problem involves finding a *Hamiltonian cycle* of minimum total length in a complete, weighted graph. Eventually, the requirement of a complete graph can be relaxed. The TSP is a NP-hard problem.

Consider the following graph, draw in python with the package **networkx**.

```python
#Simple graph implementation

import networkx as nx
import numpy as np
import random

G = nx.Graph()
G = nx.Graph()
G.add_node(1,pos=(0,0))
G.add_node(2,pos=(0,1))
G.add_node(3,pos=(0,2))
G.add_node(4,pos=(0,3))
G.add_node(5,pos=(1,0))
G.add_node(6,pos=(1,1))
G.add_node(7,pos=(1,2))
G.add_node(8,pos=(1,3))

G.add_edge(1,2,weight=1)
G.add_edge(1,3,weight=2)
G.add_edge(1,4,weight=3)
G.add_edge(1,5,weight=1)
G.add_edge(1,6,weight=np.sqrt(2))
G.add_edge(1,7,weight=np.sqrt(5))
G.add_edge(1,8,weight=np.sqrt(10))
G.add_edge(2,3,weight=1)
G.add_edge(2,4,weight=2)
G.add_edge(2,5,weight=np.sqrt(2))
G.add_edge(2,6,weight=1)
G.add_edge(2,7,weight=np.sqrt(2))
G.add_edge(2,8,weight=np.sqrt(5))
G.add_edge(3,4,weight=1)
G.add_edge(3,5,weight=np.sqrt(5))
G.add_edge(3,6,weight=np.sqrt(2))
G.add_edge(3,7,weight=1)
G.add_edge(3,8,weight=np.sqrt(2))
G.add_edge(4,5,weight=np.sqrt(10))
G.add_edge(4,6,weight=np.sqrt(5))
G.add_edge(4,7,weight=np.sqrt(2))
G.add_edge(4,8,weight=1)
G.add_edge(5,6,weight=1)
G.add_edge(5,7,weight=2)
G.add_edge(5,8,weight=3)
G.add_edge(6,7,weight=1)
G.add_edge(6,8,weight=2)
G.add_edge(7,8,weight=1)

pos=nx.get_node_attributes(G,'pos')
nx.draw(G,pos,with_labels = True)
```

### 3. Cost function

Represent your tour by a permutation of the $n$ cities, indicating the order in which the cities are visited.

The networkx package uses a representation with n+1 nodes (node n+1 is equal to node 1). Assume your christofides solution is in list Chr, you can remove the last node with the following code, so that your solution is a permutation of the cities.

> del Chr[-1]

Define a function that returns the cost of a tour. Use this function to compute the cost of your Christofides solution. You will need the following networkx function to get the weight of edge (i,j):

> G[i][j]['weight']

```python
#Answer
def cost(tour,G):
    tourCost = 0
    for i in range(0,len(tour)-1):
        tourCost += G[tour[i]][tour[i+1]]["weight"]
    return tourCost + G[tour[len(tour)-1]][tour[0]]["weight"]

print(Chr)
cost(Chr,G)
```

### 4. Nearest neighbour

Define a function that implements the nearest neighbour heuristic, starting at node 'start'.

networkx functions:

> nx.neighbors(G, i) 

returns all neighbours of node i in graph G.

```python
#ANSWER
def NN(H,start):
    G = H.copy()
    path = [start]
    cost = 0
    n = G.number_of_nodes()
    for i in range(n-1):
        distances = {d: G[path[i]][d]['weight'] for d in nx.neighbors(G, path[i])}
        nn = min(distances, key=distances.get)
        path.append(nn)
        G.remove_node(path[i])
    return(path)
    
NNtour= NN(G,1)
print(cost(NNtour,G))
```

### 5. Randomised nearest neighbour

Modify your function in question 5 to create a greedy randomised NN heuristic

```python
#ANSWER
def NNRand(H,start,alpha):
    G = H.copy()
    path = [start]
    cost = 0
    n = G.number_of_nodes()
    for i in range(n-1):
        distances = {d: G[path[i]][d]['weight'] for d in nx.neighbors(G, path[i])}
        sort_distances = sorted(distances.items(), key=lambda x: x[1])
        selected = random.randint(0, min(alpha,n-i-2))
        nn = sort_distances[selected][0] 
        path.append(nn)
        G.remove_node(path[i])
    return(path)

NNtour= NNRand(G,1,2)
print(NNtour)
print(cost(NNtour,G))
```

###  6. Representation for constrained spanning tree problems.  (Talbi)

Given a connected graph $G=(V, E)$, a spanning tree is a minimum size connected and maximum size acyclic subgraph of $G$ spanning all the vertices of $V$. The large numbers of applications have required the study of variants of the well-known minimum spanning tree problem (MSTP). Given a connected graph $G=(V, E)$, with $n=|V|, m=|E|$, a spanning tree is a connected and acyclic subgraph of $G$ spanning all the vertices of $V$ with $n-1$ edges. Although the MSTP, the more studied problem involving spanning tree, can be solved in polynomial time, the outstanding importance of spanning trees in telecommunication or integrated circuit network design, biology, or computer science has required the development of more complex problems and often NP-hard variants. Indeed, adding some constraints (e.g., node degree, graph diameter) to the MSTP problem makes it NP-hard.

For instance, in the hop-constrained minimum spanning tree problem (HMSTP), the unique path from a specified root node, node 0 , to any other node has no more than $H$ hops (edges). Propose an encoding for the HMSTP problem.

#### Answer

As in any other problem, there are multiple possible encodings. Each one with its owns advantages and problems. Below are two options.

#### Option 1) 

One could think of an incomplete representation based on random keys. Consider a vector of random keys where each element of the vector correspond to the weight associated with one of the edges.

Decoding procedure:

- create a MST
- For each node, evaluate its 'hop'.
- For nodes that violate the hop constraint, connect them to the closest node such that the hop constraint is no longer violated.

Note that the decoding procedure above may fail if the graph is not complete.



#### Option 2)

A vector of integers with each position associated with one node. The value in the position indicates the parent of the associated node. This representation has a large number of infeasible realisations.

### 7. Greedy algorithms for vehicle routing problems (Talbi)
 
Vehicle routing problems represent very important applications in the area of logistics and transportation. VRP are some of the most studied problems in the combinatorial optimization domain. The problem was introduced more than four decades ago by Dantzig and Ramser. The basic variant of the VRP is the capacitated vehicle routing problem. CVRP can be defined as follows: Let $G=(V, A)$ be a graph where $V$ the set of vertices represents the customers. One vertex represents the depot with a fleet of $m$ identical vehicles of capacity $Q$. We associate with each customer $v_i$ a demand $q_i$ and with each edge $\left(v_i, v_j\right)$ of $A$ a cost $c_{i j}$. We have to find a set of routes where the objective is to minimize the total cost and satisfy the following constraints:
- For each vehicle, the total demand of the assigned customers does not exceed its capacity $Q$.
- Each route must begin and end at the depot node. 
- Each customer is visited exactly once.

Define one or more greedy algorithms for the CVRP problem.

#### Answer

For example, we can use two classical heuristics known as "The Clarke and Wright savings heuristic" and "The sweep algorithm". The descriptions below are from Cordeau, J.-F., Gendreau, M., Laporte, G., Potvin, J.-Y., & Semet, F. (2002). A guide to vehicle routing heuristics. Journal of the Operational Research Society, 53(5), 512–522. https://doi.org/10.1057/palgrave.jors.2601319


#### The Clarke and Wright savings heuristic

The Clarke and Wright $(\mathrm{CW})$ heuristic is one of the best known and remains widely used in practice to this day, despite some of its shortcomings. It is based on the notion of saving. Initially, a feasible solution consists of $n$ back and forth routes between the depot and a customer. At any given iteration, two routes $\left(v_0, \ldots, v_i, v_0\right)$ and $\left(v_0, v_j, \ldots, v_0\right)$ are merged into a single route $\left(v_0, \ldots, v_i, v_j \ldots, v_0\right)$ whenever this is feasible, thus generating a saving $s_{i j}=c_{i 0}+c_{0 j}-c_{i j}$. In the parallel version of the algorithm, the merge yielding the largest saving is always implemented, whereas the sequential version keeps expanding the same route until this is no longer feasible. In practice the parallel version is much better. It is common to apply a 3 -opt post-optimization step to the final solution.


#### The sweep algorithm

The sweep algorithm is generally attributed to Gillett and Miller although its principle can be traced back to Wren and Holliday. It applies to planar instances of the VRP. Feasible routes are created by rotating a ray centred at the depot and gradually including customers in a vehicle route until the capacity or route length constraint is attained. A new route is then initiated and the process is repeated until the entire plane has been swept. A 3-opt step is then typically applied.
